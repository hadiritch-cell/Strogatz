```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson

# --------------------------------------
# 1. Sample and plot ER networks for p=0.000001, 0.001, 0.01, 0.05
# --------------------------------------

n = 1000
p_values = [0.000001, 0.001, 0.01, 0.05]

plt.figure(figsize=(18,4)) # Adjusted figsize to accommodate 4 subplots

for i, p in enumerate(p_values):
  G = nx.erdos_renyi_graph(n, p)

  plt.subplot(1, len(p_values), i+1) # Changed subplot columns to match number of p_values
  nx.draw_spring(G, node_size=15, node_color="red",edge_color="grey", alpha=0.05)
  plt.title(f"ER Graph n={n}, p={p}")

plt.tight_layout()
plt.show()
```
